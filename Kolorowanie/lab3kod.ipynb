```python
#Kolorowanie składni języka python 

def scanner(expression, position):
    keywords1 = {'class', 'for', 'while', 'if', 'else', 'elif', 'def', 'return', 'print', 'pass'}
    keywords2 = {'or', 'and', 'range', 'in', 'break', 'continue', 'True', 'False'}
    token = None
    value = ''
    character = expression[position]
    length = len(expression)

    if character.isdigit():
        while position < length and expression[position].isdigit():
            value += expression[position]
            position += 1
        token = 'INTEGER'

    elif character.isalpha():
        while position < length and expression[position].isalnum():
            value += expression[position]
            position += 1
        if value in keywords1 or value in keywords2:
            token = value.upper()
        else:
            token = 'IDENTIFIER'
    elif character == '#':  # Handle comments
        while position < length and expression[position] != '\n':
            value += expression[position]
            position += 1
        token = 'COMMENT'
    elif character == '"':  # Handle string literals
        value += character
        position += 1
        while position < length and expression[position] != '"':
            value += expression[position]
            position += 1
        if position < length:
            value += expression[position]  # Dodaj znak cudzysłowu kończącego łańcuch
            position += 1
        token = 'STRING'
    elif character == '[':  # Handle lists
        value += character
        position += 1
        while position < length and expression[position] != ']':
            value += expression[position]
            position += 1
        if position < length:
            value += expression[position]  # Add the closing square bracket to the list
            position += 1
        token = 'LIST'
    

    else:
        if character in '([{':
            token = 'LEFT_PARENTHESIS'
        elif character in ')]}':
            token = 'RIGHT_PARENTHESIS'
        elif character in '+-*/=':
            token = 'OPERATOR'
        elif character in '<>':
            token = 'COMPARISON'
        elif character == ':':
            token = 'COLON'
        elif character == ' ':
            token = 'SPACE'
        elif character == '\n':
            token = "NEW_LINE"
        elif character == '"':  # Handle string literals
            token = 'STRING'
        elif character in '+-*/=':
            token = 'OPERATOR'
        
            
    position += 1

    if token:
        return token, value if value else character
    else:
        return 'Error', f'Nieznany token na pozycji: {position}'
```

```python
#kolorowanie
def apply_coloring(input_file, output_file):
    with open(input_file, 'r') as f:
        code = f.read()

    position = 0
    html_code = '<pre>'
    while position < len(code):
        token, value = scanner(code, position)
        if token == 'INTEGER':
            html_code += f'<span style="color: blue">{value}</span>'
        elif token == 'IDENTIFIER':
            html_code += f'<span style="color: green">{value}</span>'
        elif token in ('LEFT_PARENTHESIS', 'RIGHT_PARENTHESIS'):
            html_code += f'<span style="color: #FFCC00">{value}</span>'
        elif token in ('OPERATOR', 'COMPARISON'):
            html_code += f'<span style="color: red">{value}</span>'
        elif token in ('CLASS', 'FOR', 'WHILE', 'IF', 'ELSE', 'ELIF', 'DEF', 'RETURN', 'PRINT', 'PASS'):
            html_code += f'<span style="color: #EA59F2">{value}</span>'
        elif token in ('IN', 'RANGE', 'AND', 'OR', 'BREAK', 'CONTINUE', 'TRUE', 'FALSE'):
            html_code += f'<span style="color: #38B0DE">{value}</span>'
        elif token == 'COLON':
            html_code += ':'
        elif token == 'SPACE':
            html_code += ' '
        elif token == 'NEW_LINE':
            html_code += '\n'
        elif token == 'COMMENT':
            html_code += f'<span style="color: grey">{value}</span>'
        elif token == 'STRING':
            html_code += f'<span style="color: #A38585">{value}</span>' 
        elif token == 'LIST':
            html_code += f'<span style="color: #54B7B1">{value}</span>'
        else:   
            html_code += value
        position += len(value)
        
    html_code += '</pre>'

    with open(output_file, 'w') as f:
        f.write(html_code)
```

```python
apply_coloring('input_code.txt', 'output_code.html')
```
